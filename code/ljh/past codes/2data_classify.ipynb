```python
import os

import pandas as pd

DATA_PATH = '/opt/ml/input/cropped_v2/train/'
IMG_PATH = '/opt/ml/input/cropped_v2/train/images/'
```

```python
df = pd.read_csv('/opt/ml/input/cropped_v2/train/train_outlier.csv')
```

```python
img_path_list = []
```

```python
for p in df['path'].values:
    try : img_list = sorted([im for im in os.listdir(IMG_PATH+p) if '._' not in im])
    except : pass
    for im in img_list:
        img_path_list.append(p+'/'+im)
```

```python
len(img_path_list)
```

```python
df_cls = pd.DataFrame( { 'img_path' : img_path_list} )
```

```python
def path_finder(x):
    return x.split('/')[0]
```

```python
def name_finder(x):
    return x.split('/')[-1]
```

```python
def mask_classifier(x):
    mask = x.split('.')[0]
    
    if mask == 'incorrect_mask':
        return 'Incorrect'
    elif mask == 'normal':
        return 'Not Wear'
    else:
        return 'Wear'
```

```python
df_cls['path'] = df_cls['img_path'].map(path_finder)
df_cls['name'] = df_cls['img_path'].map(name_finder)
df_cls['mask'] = df_cls['name'].map(mask_classifier)
```

```python
df_cls = pd.merge(df, df_cls, on='path')
```

```python
df_cls['target'] = 0


df_cls['target'] = ((df_cls['mask']=='Not Wear')*12 + 
                  (df_cls['mask']=='Incorrect')*6 +
                  (df_cls['gender']=='female')*3 + 
                  (df_cls['age']>=30)*1 + 
                  (df_cls['age']>=60)*1)
```
